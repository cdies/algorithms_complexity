```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.DataFrame(np.arange(50))
df.columns = ["NUM"]
```

```python
df["CONST"] = 1
```

```python
df["LOG_N"] = np.log(df.NUM)
df.LOG_N.iat[0] = 0
```

```python
df["N_LOG_N"] = df.LOG_N * df.NUM
```

```python
df["N_2"] = df.NUM ** 2
df["N_3"] = df.NUM ** 3
```

```python
df.head()
```

```python
sns.set(rc={"figure.figsize":(7, 7)})
```

```python
fig, ax = plt.subplots()

sns.lineplot(x="NUM", y="CONST", data=df, label="Constanta", linewidth=3.0)
sns.lineplot(x="NUM", y="LOG_N", data=df, label="Log(N)", linewidth=3.0)
sns.lineplot(x="NUM", y="NUM", data=df, label="N", linewidth=3.0)
sns.lineplot(x="NUM", y="N_LOG_N", data=df, label="N*Log(N)", linewidth=3.0)
sns.lineplot(x="NUM", y="N_2", data=df, label="N**2", linewidth=3.0)
sns.lineplot(x="NUM", y="N_3", data=df, label="N**3", linewidth=3.0)

ax.set_ylim(0, 50)
ax.set_xlabel("Количество элементов")
ax.set_ylabel("Время (количество операций)")
plt.show()
```
